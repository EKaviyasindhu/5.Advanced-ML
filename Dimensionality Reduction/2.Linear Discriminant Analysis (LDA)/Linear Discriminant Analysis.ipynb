```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
# ---------------------------
# Functions
# ---------------------------
def Func_LDA(X_train, y_train, X_test):
    lda = LDA(n_components=2)
    X_train_lda = lda.fit_transform(X_train, y_train)
    X_test_lda = lda.transform(X_test)
    return lda, X_train_lda, X_test_lda

def evaluate_model(name, classifier, X_train, y_train, X_test, y_test):
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    print(f"\n=== {name} ===")
    print("Accuracy:", acc)
    print("Confusion Matrix:\n", cm)
    return acc, cm
```

```python
# ---------------------------
# Load dataset
# ---------------------------
dataset = pd.read_csv("Wine.csv")
X = dataset.iloc[:, 0:13].values
y = dataset.iloc[:, 13].values

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Scale
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Apply LDA
lda, X_train_lda, X_test_lda = Func_LDA(X_train, y_train, X_test)
```

```python
# ---------------------------
# Train classifiers on LDA data
# ---------------------------
results = {}
results["Logistic"] = evaluate_model("Logistic Regression", LogisticRegression(random_state=0), X_train_lda, y_train, X_test_lda, y_test)
results["SVM (linear)"] = evaluate_model("SVM (linear)", SVC(kernel='linear', random_state=0), X_train_lda, y_train, X_test_lda, y_test)
results["SVM (rbf)"] = evaluate_model("SVM (rbf)", SVC(kernel='rbf', random_state=0), X_train_lda, y_train, X_test_lda, y_test)
results["KNN"] = evaluate_model("KNN", KNeighborsClassifier(n_neighbors=5), X_train_lda, y_train, X_test_lda, y_test)
results["Naive Bayes"] = evaluate_model("Naive Bayes", GaussianNB(), X_train_lda, y_train, X_test_lda, y_test)
results["Decision Tree"] = evaluate_model("Decision Tree", DecisionTreeClassifier(criterion='entropy', random_state=0), X_train_lda, y_train, X_test_lda, y_test)
results["Random Forest"] = evaluate_model("Random Forest", RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0), X_train_lda, y_train, X_test_lda, y_test)

# ---------------------------
# Build summary table
# ---------------------------
summary = {name: acc for name, (acc, cm) in results.items()}
result_df = pd.DataFrame(summary, index=["PCA"])
print("\n=== Summary Table (PCA) ===")
print(result_df)
```
